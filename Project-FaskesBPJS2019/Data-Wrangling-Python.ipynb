Import Module dan Dataset

```python
import pandas as pd
import numpy as np
import re

csv_data = pd.read_csv("Data Faskes BPJS 2019.csv")
```

Merapikan Kolom KotaKab, NamaFaskes, dan TelpFaskes

```python
def delspace(row):
    cln = " ".join(row.split())
    return cln
def delWord(row):
    cln = row.replace('Kode Faskes dan Alamat Rumah Sakit BPJS di ','')
    return cln
```

```python
csv_data['KotaKab'] = csv_data['KotaKab'].apply(delspace)
csv_data['KotaKab'] = csv_data['KotaKab'].apply(delWord)
csv_data['KotaKab']
```

```python
csv_data['NamaFaskes'] = csv_data['NamaFaskes'].apply(delspace)
csv_data['NamaFaskes']
```

```python
csv_data['TelpFaskes'] = csv_data['TelpFaskes'].apply(delspace)
csv_data['TelpFaskes']
```

Menghapus kolom NoLink

```python
csv_data.drop('NoLink', inplace=True, axis=1)
```

Mengganti strip dengan NaN

```python
csv_data = csv_data.astype(str)
data = csv_data.applymap(lambda x: re.sub(r'^-$', str(np.NaN), x))
csv_data
```

Menyimpan data ke file baru

```python
csv_data.to_csv("Bersih Data Faskes BPJS 2019.csv", index = False)
```
